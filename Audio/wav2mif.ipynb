```python
from scipy.io.wavfile import read
import scipy.signal as sps

# filename = "gameover"
Fs, data = read("contra.wav")

# Resample data

new_Fs = 8000 # new sampling rate
number_of_samples = round(len(data) * float(new_Fs) / Fs)
data = sps.resample(data, number_of_samples)

depth = data.shape[0] + 1

outFile = open("contra_hex_rom.mif", 'w')
outFile.write("WIDTH = 16;\n")
outFile.write("DEPTH = %d;\n" %depth)
outFile.write("ADDRESS_RADIX = HEX;\n")
outFile.write("DATA_RADIX = HEX;\n")
outFile.write("CONTENT BEGIN\n")
outFile.write("\n")

for i in range(data.shape[0]):
    outFile.write("%X: %04X;\n" %(i+1, ( int(round(data[i,0])) & 0xffff )))

outFile.write("END;\n")
outFile.close()
```
